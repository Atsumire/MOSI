# Применить к текстам лемматизацию, удаление стоп слов и токенизацию по словам

## Токенизация по словам

```python
def Tokenization(text):
    text = text.lower()
    word_tokens = word_tokenize(text)
    return word_tokens
```

### Русский

```python
tokenTextRu = Tokenization(textRu)
print(tokenTextRu)
```

### Английский

```python
tokenTextEng = Tokenization(textEn)
print(tokenTextEng)
```

## Лемматизация

```python
def Lemmatized(text, lang):
    morph = pymorphy3.MorphAnalyzer()
    if lang != 'ru' and lang != 'eng': return 'error'
    tokens = Tokenization(text)
    lemmas = [morph.parse(token)[0].normal_form for token in tokens]
    return lemmas
```

### Русский

```python
lemmaTextRu = Lemmatized(textRu, 'ru')
print(lemmaTextRu)
```

### Английский

```python
lemmaTextEng = Lemmatized(textEn, 'eng')
print(lemmaTextEng)
```

## Удаление стоп слов

```python
def delStopWords(tokens, lang):
    if lang == 'ru': stop_words = set(stopwords.words('russian'))
    elif lang == 'eng': stop_words = set(stopwords.words('english'))
    else: return
    filtered_tokens = [word for word in tokens if word not in stop_words]
    return filtered_tokens
```

### Русский

```python
noStopTextRu = delStopWords(lemmaTextRu, 'ru')
print(noStopTextRu)
```

### Английский

```python
noStopTextEng = delStopWords(lemmaTextEng, 'eng')
print(noStopTextEng)
```

# Реализовать Bag of Words

### Разбить текст на предложения

```python
def splitText(text):
    texts = list(set(text.split('\n')))
    if '' in texts: texts.remove('')
    return texts
```

```python
def BagOfWords(text, lang):
    texts = splitText(text)
    tokens_bag = list()
    words = set()
    
    for text_i in texts:
        tokens = Lemmatized(text_i, lang)
        tokens = delStopWords(tokens, lang)
        
        tokens_bag.append(tokens)
        words.update(tokens)
    
    words = list(words)
    matrix = np.zeros((len(texts), len(words)), dtype=int)
    i=0
    
    for tokens in tokens_bag:
        for token in tokens:
            if token in words:
                matrix[i, words.index(token)] += 1
        i+=1
    return matrix, words
```

### Русский

```python
matrix, feature_names = BagOfWords(textRu, 'ru')
print(feature_names)
print_matrix(matrix)
```

### Английский

```python
matrix, feature_names = BagOfWords(textEn, 'eng')
print(feature_names)
print_matrix(matrix)
```

# Реализовать TF-IDF

```python
def TF(texts, words, lang):
    tf_matrix = []
    for text in texts:
        tokens = Lemmatized(text, lang)
        tokens = delStopWords(tokens, lang)
        
        counts = [0] * len(words)
        for token in tokens:
            if token in words:
                counts[words.index(token)] += 1
                
        total_words = len(tokens)
        tf_vector = [count / total_words for count in counts]
        tf_matrix.append(tf_vector)
    return np.array(tf_matrix)

def IDF(texts, words):
    idf_vector = []

    for word in words:
        count = sum(1 for text in texts if word in text) #возвращает количество документов, в которых присутствует данное слово. Если слово не встречается ни в одном документе, count будет равно 0.
        idf = math.log(len(texts) / (count + 1))  # добавляем 1 для избежания деления на ноль
        idf_vector.append(idf)

    return idf_vector

def TFIDF(text, lang):
    words = list()
    texts = splitText(text)
    for text in texts:
        tokens = Lemmatized(text, lang)
        tokens = delStopWords(tokens, lang)
        for token in tokens: words.append(token)
    words = list(set(words))
    
    tf_matrix = TF(texts, words, lang)
    idf_vector = IDF(texts, words)
    tfidf_matrix = tf_matrix * idf_vector
    
    return tfidf_matrix, words
```

### Русский

```python
matrix, feature_names = TFIDF(textRu, 'ru')
print(feature_names)
print_matrix(matrix)
```

### Английский

```python
matrix, feature_names = TFIDF(textEn, 'eng')
print(feature_names)
print_matrix(matrix)
```
