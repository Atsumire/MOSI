# Применить к текстам лемматизацию, удаление стоп слов и токенизацию по словам

## Лемматизация

```python
def Lemmatized(text, lang):
    morph = pymorphy3.MorphAnalyzer()
    if lang == 'ru': tokens = nltk.word_tokenize(text, language='russian')
    elif lang == 'eng': tokens = nltk.word_tokenize(text, language='english')
    else: return 'error'
    lemmas = [morph.parse(token)[0].normal_form for token in tokens]
    return lemmas
```

### Русский

```python
lemmaTextRu = Lemmatized(textRu, 'ru')
print(lemmaTextRu)
```

### Английский

```python
lemmaTextEng = Lemmatized(textEn, 'eng')
print(lemmaTextEng)
```

## Удаление стоп слов

```python
def delStopWords(text, lang):
    tokens = word_tokenize(text)
    if lang == 'ru': stop_words = set(stopwords.words('russian'))
    elif lang == 'eng': stop_words = set(stopwords.words('english'))
    else: return
    filtered_tokens = [word for word in tokens if word not in stop_words]
    return filtered_tokens
```

### Русский

```python
noStopTextRu = delStopWords(textRu, 'ru')
print(noStopTextRu)
```

### Английский

```python
noStopTextEng = delStopWords(textEn, 'eng')
print(noStopTextEng)
```

## Токенизация по словам

```python
def Tokenization(text, lang):
    text = text.lower()
    word_tokens = word_tokenize(text)
    return word_tokens
```

### Русский

```python
tokenTextRu = Tokenization(textRu, 'ru')
print(tokenTextRu)
```

### Английский

```python
tokenTextEng = Tokenization(textEn, 'eng')
print(tokenTextEng)
```

# Реализовать Bag of Words

```python
def BagOfWords(text, lang):
    vectorizer = CountVectorizer()
    tokens = Tokenization(text, lang)
    matrix = vectorizer.fit_transform(tokens)
    return matrix.toarray(), vectorizer.get_feature_names_out()
```

### Русский

```python
matrix, feature_names = BagOfWords(textRu, 'ru')
print(feature_names)
print_matrix(matrix)
```

### Английский

```python
matrix, feature_names = BagOfWords(textEn, 'eng')
print(feature_names)
print_matrix(matrix)
```

# Реализовать TF-IDF

```python
def TFIDF(text, lang):
    vectorizer = TfidfVectorizer()
    tokens = Tokenization(text, lang)
    matrix = vectorizer.fit_transform(tokens)
    return matrix.toarray(), vectorizer.get_feature_names_out()
```

### Русский

```python
matrix, feature_names = TFIDF(textRu, 'ru')
print(feature_names)
print_matrix(matrix)
```

### Английский

```python
matrix, feature_names = TFIDF(textEn, 'eng')
print(feature_names)
print_matrix(matrix)
```
